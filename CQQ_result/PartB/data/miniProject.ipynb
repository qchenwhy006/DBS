```python
import operator
import lightgbm as lgb
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
import seaborn as sns
import xgboost as xgb
from sklearn import metrics
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
import tensorflow.keras.backend as K
from sklearn.preprocessing import LabelBinarizer, StandardScaler
```

```python
Multi_labels = [
    "soul and reggae",
    "pop",
    "punk",
    "jazz and blues",
    "dance and electronica",
    "folk",
    "classic pop and rock",
    "metal"
]

Binary_labels = [
    "soul and reggae",
    "punk"        
]

MultiLabel_to_id = {label: idx for idx, label in enumerate(Multi_labels)}
id_to_MultiLabel = {idx: label for idx, label in enumerate(Multi_labels)}
BinaryLabel_to_id = {label: idx for idx, label in enumerate(Binary_labels)}
id_to_BinaryLabel = {idx: label for idx, label in enumerate(Binary_labels)}


def build_dataset(tf_or_pd, multi_or_binary):
    """
    Preprocess and build the dataset for machine learning and deep learning model
    :return: input and labels related to train and test dataset
    """

    feature_df = pd.read_csv('features.csv')
    label_df = pd.read_csv('labels.csv')
    test_df = pd.read_csv('test.csv')

    df = pd.concat([feature_df, label_df['genre']], axis=1)
    
    if multi_or_binary == 'binary':
        df = df.loc[operator.or_(df['genre'] == 'soul and reggae', df['genre'] == 'punk')]      
    df = df.reset_index(drop=True)

    df.fillna(df.median(),inplace=True)
    df_test = test_df.dropna(axis=0, how='any')
    print('There are total {} data removed from test dataset due to the missing value'.format(
        len(test_df) - len(df_test)))


    removed_list = ['trackID', 'title', 'tags', 'duration']
    print('There are total {} columns removed due to the text property'.format(len(removed_list)))
    df_train = df.drop(removed_list, axis=1)
    df_test = df_test.drop(removed_list, axis=1)
    
    if tf_or_pd == 'tf':
        one_hot = LabelBinarizer()
        train_y = one_hot.fit_transform(df_train['genre'])

        ss = StandardScaler()
        train_x = ss.fit_transform(df_train.iloc[:, :-1])
        test_x = ss.fit_transform(df_test)
    else:
        train_x = df_train.iloc[:, :-1]
        if multi_or_binary == 'multi':
            train_y = df_train.iloc[:, -1].apply(lambda x: MultiLabel_to_id[x])
        else:
            train_y = df_train.iloc[:, -1].apply(lambda x: BinaryLabel_to_id[x])
        test_x = df_test

    return train_x, train_y, test_x, one_hot if tf_or_pd == 'tf' else None
```

```python
def train_model_classification(X, X_test, y, params, num_classes,
                               folds=None, model_type='lgb',
                               eval_metric='logloss',
                               plot_feature_importance=True,
                               verbose=10000,
                               early_stopping_rounds=200,
                               splits=None, n_folds=3):
    """
    Multi_classes classifier

    :params: X - input for training
    :params: X_test - input for testing
    :params: y - labels of train dataset
    :params: params - parameters for different models
    :params: num_classes - the number of classes required to be classified
    :params: folds - folds to split data
    :params: model_type - Lightgbm or Xgboost
    :params: eval_metric - evaluation metric and loss function
    :params: columns - the number of feature
    :params: plot_feature_importance - whether plot the importance of features

    """
    global y_pred_valid, y_pred

    X = pd.DataFrame(X)
    X_test = pd.DataFrame(X_test)
    y = pd.DataFrame(y)

    columns = X.columns
    splits = folds.split(X, y) if splits is None else splits
    n_splits = folds.n_splits if splits is None else n_folds

    metrics_dict = {
        'logloss': {
            'lgb_metric_name': 'logloss',
            'xgb_metric_name': 'mlogloss',
            'sklearn_scoring_function': metrics.log_loss
        },
        'lb_score_method': {
            'sklearn_scoring_f1': metrics.f1_score,
            'sklearn_scoring_accuracy': metrics.accuracy_score,
            'sklearn_scoring_auc': metrics.roc_auc_score
        },
    }
    result_dict = {}

    oof = np.zeros(shape=(len(X), num_classes))
    prediction = np.zeros(shape=(len(X_test), num_classes))
    scores = []
    accuracy = []
    feature_importance = pd.DataFrame()

    for fold_n, (train_index, valid_index) in enumerate(splits):
        if type(X) == np.ndarray:
            X_train, X_valid = X[train_index], X[valid_index]
            y_train, y_valid = y[train_index], y[valid_index]
        else:
            X_train, X_valid = X[columns].iloc[train_index], X[columns].iloc[valid_index]
            y_train, y_valid = y.iloc[train_index], y.iloc[valid_index]

        if model_type == 'lgb':
            model = lgb.LGBMClassifier(**params)
            model.fit(X_train, y_train,
                      eval_set=[(X_train, y_train), (X_valid, y_valid)],
                      eval_metric=metrics_dict[eval_metric]['lgb_metric_name'],
                      verbose=verbose,
                      early_stopping_rounds=early_stopping_rounds)

            y_pred_valid = model.predict_proba(X_valid)
            y_pred = model.predict_proba(X_test, num_iteration=model.best_iteration_)

        if model_type == 'xgb':
            model = xgb.XGBClassifier(**params)
            model.fit(X_train, y_train,
                      eval_set=[(X_train, y_train), (X_valid, y_valid)],
                      eval_metric=metrics_dict[eval_metric]['xgb_metric_name'],
                      verbose=bool(verbose),  # xgb verbose bool
                      early_stopping_rounds=early_stopping_rounds)
            y_pred_valid = model.predict_proba(X_valid)
            y_pred = model.predict_proba(X_test, ntree_limit=model.best_ntree_limit)

        oof[valid_index] = y_pred_valid
        scores.append(metrics_dict['lb_score_method']['sklearn_scoring_f1'](y_valid, np.argmax(y_pred_valid, axis=1),
                                                                            average='macro'))
        accuracy.append(metrics_dict['lb_score_method']['sklearn_scoring_accuracy'](y_valid, np.argmax(y_pred_valid, axis=1)))
        prediction += y_pred

        if model_type == 'lgb' and plot_feature_importance:
            # feature importance
            fold_importance = pd.DataFrame()
            fold_importance["feature"] = columns
            fold_importance["importance"] = model.feature_importances_
            fold_importance["fold"] = fold_n + 1
            feature_importance = pd.concat([feature_importance, fold_importance], axis=0)

        if model_type == 'xgb' and plot_feature_importance:
            # feature importance
            fold_importance = pd.DataFrame()
            fold_importance["feature"] = columns
            fold_importance["importance"] = model.feature_importances_
            fold_importance["fold"] = fold_n + 1
            feature_importance = pd.concat([feature_importance, fold_importance], axis=0)
    prediction /= n_splits
    print('CV mean score: {0:.4f}, std: {1:.4f}.'.format(np.mean(scores), np.std(scores)))

    result_dict['oof'] = oof
    result_dict['prediction'] = prediction
    result_dict['scores'] = scores
    result_dict['accuracy'] = accuracy

    if model_type == 'lgb' or model_type == 'xgb':
        if plot_feature_importance:
            feature_importance["importance"] /= n_splits
            cols = feature_importance[["feature", "importance"]].groupby("feature").mean().sort_values(
                by="importance", ascending=False)[:50].index

            best_features = feature_importance.loc[feature_importance.feature.isin(cols)]

            plt.figure(figsize=(16, 12))
            sns.barplot(x="importance", y="feature", data=best_features.sort_values(by="importance", ascending=False))
            plt.title('{} Features (avg over folds)'.format(model_type))
            plt.show()
            result_dict['feature_importance'] = feature_importance

    print("Accuracy of our model on test data : ", sum(result_dict['accuracy'])/len(result_dict['accuracy']) * 100, "%")
    
    if num_classes == 2:
        print(y_valid)
        draw_confusion_matrix(y_valid, y_pred_valid, _, 'pd', 'binary')
    else:
        draw_confusion_matrix(y_valid, y_pred_valid, _, 'pd', 'multi')
    return result_dict, model
```

```python
def draw_confusion_matrix(y_valid, y_pred_valid, one_hot, tf_or_pd, binary_or_multiclass):
    if tf_or_pd == 'pd':
        y_valid = pd.DataFrame(y_valid, columns=['genre'])
        y_pred_valid = pd.DataFrame(np.argmax(y_pred_valid, axis=1), columns=['genre'])
        
        if binary_or_multiclass == 'multi':
            y_valid = np.array(y_valid['genre'].apply(lambda x: id_to_MultiLabel[x]))
            y_pred_valid = np.array(y_pred_valid['genre'].apply(lambda x: id_to_MultiLabel[x]))
        else:
            
            y_valid = np.array(y_valid['genre'].apply(lambda x: id_to_BinaryLabel[x]))
            y_pred_valid = np.array(y_pred_valid['genre'].apply(lambda x: id_to_BinaryLabel[x]))
            
    else:
        y_pred_valid = one_hot.inverse_transform(y_pred_valid)
        y_valid = one_hot.inverse_transform(y_val.numpy())
        
    cm = confusion_matrix(y_valid, y_pred_valid)
    plt.figure(figsize=(12, 10))
    
    if binary_or_multiclass == 'multi':
        cm = pd.DataFrame(cm, index=[i for i in Multi_labels], columns=[i for i in Multi_labels])
    else:
        cm = pd.DataFrame(cm, index=[i for i in Binary_labels], columns=[i for i in Binary_labels])
    
    sns.heatmap(cm, linecolor='white', cmap='Blues', linewidth=1, annot=True, fmt='')
    plt.title('Confusion Matrix', size=20)
    plt.xlabel('Predicted Labels', size=14)
    plt.ylabel('Actual Labels', size=14)
    plt.show()
    if binary_or_multiclass == 'multi':
        print(classification_report(y_valid, y_pred_valid, target_names = Multi_labels))
    else:
        print(classification_report(y_valid, y_pred_valid, target_names = Binary_labels))
```

# DeepLearning Method

```python
X, y, X_test, one_hot = build_dataset('tf', 'multi')
num_classes = 8
column_num = len(X[0])
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=0, shuffle=True, test_size=0.2, stratify=y)
epochs = 200
batch_size = 32
```

```python
X_train = tf.convert_to_tensor(np.array(X_train), dtype=tf.float64)
X_val = tf.convert_to_tensor(np.array(X_val), dtype=tf.float64)
y_train = tf.convert_to_tensor(np.array(y_train), dtype=tf.float64)
y_val = tf.convert_to_tensor(np.array(y_val), dtype=tf.float64)
```

```python
def Encoder(tf_x):
    en0 = tf.keras.layers.Dense(256, tf.nn.sigmoid)(tf_x)
    en1 = tf.keras.layers.Dense(128, tf.nn.sigmoid)(en0)
    en2 = tf.keras.layers.Dense(128, tf.nn.sigmoid)(en1)
    encoded = tf.keras.layers.Dense(128, tf.nn.sigmoid, name='Encoder_layer')(en2)
    return encoded

def Encoder_Classifier(column_num, num_classes):
    input = tf.keras.Input(shape=(column_num,))
    encoded_input = Encoder(input)
    X_flatten = tf.keras.layers.Flatten()(encoded_input)
    y = tf.keras.layers.Dense(num_classes, activation='softmax')(X_flatten)
    model = tf.keras.Model(input, y)
    return model

def micro_f1(y_true, y_pred):
    TP = K.cast(K.sum(K.round(K.clip(y_true*y_pred, 0, 1)), axis=0), tf.float64)
    predicted_positives = K.cast(K.sum(K.round(K.clip(y_pred, 0, 1)), axis=0), tf.float64)
    real_positives = K.cast(K.sum(K.round(K.clip(y_true, 0, 1)), axis=0), tf.float64)

    precision = K.sum(TP)/(K.sum(predicted_positives) + K.epsilon())
    recall = K.sum(TP)/(K.sum(real_positives) + K.epsilon())
    micro_f1 = 2 * precision * recall / (precision + recall + K.epsilon())

    return micro_f1


def macro_f1(y_true, y_pred):
    TP = K.cast(K.sum(K.round(K.clip(y_true * y_pred, 0, 1)), axis=0), tf.float64)
    predicted_positives =K.cast(K.sum(K.round(K.clip(y_pred, 0, 1)), axis=0), tf.float64)
    real_positives = K.cast(K.sum(K.round(K.clip(y_true, 0, 1)), axis=0), tf.float64)

    precision = TP/(predicted_positives+K.epsilon())
    recall = TP/(real_positives+K.epsilon())
    macro_f1 = K.mean(2 * precision * recall/(precision + recall+K.epsilon()))

    return macro_f1
```

```python
model = Encoder_Classifier(column_num, num_classes)
print("Building the model ...")
model.compile(tf.optimizers.Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=[micro_f1, macro_f1])
print("Starting the training ...")
history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, workers=8, use_multiprocessing=True,
                    validation_data=(X_val, y_val))

history_dict = history.history
history_dict.keys()
print("Accuracy of our model on test data : ", model.evaluate(X_val, y_val)[1] * 100, "%")

epochs = [i for i in range(epochs)]
fig, ax = plt.subplots(1, 2)
train_f1 = history.history['micro_f1']
train_loss = history.history['loss']
val_f1 = history.history['val_micro_f1']
val_loss = history.history['val_loss']

fig.set_size_inches(20, 6)
ax[0].plot(epochs, train_loss, label='Training Loss')
ax[0].plot(epochs, val_loss, label='Val Loss')
ax[0].set_title('Training & Val Loss')
ax[0].legend()
ax[0].set_xlabel("Epochs")

ax[1].plot(epochs, train_f1, label='Training Micro_F1')
ax[1].plot(epochs, val_f1, label='Val Micro_F1')
ax[1].set_title('Training & Val Micro_F1')
ax[1].legend()
ax[1].set_xlabel("Epochs")
plt.show()

y_pred_valid = model.predict(X_val)
draw_confusion_matrix(y_val, y_pred_valid, one_hot, 'tf', 'multi')
    
    
y_pred = model.predict(X_test)
y_pred = one_hot.inverse_transform(y_pred)

pred = pd.DataFrame(y_pred, columns=['genre'])   
final_result = pd.DataFrame(columns=['trackID', 'genre'])
test_df = pd.read_csv('test.csv')
final_result['trackID'] = test_df['trackID']
final_result['genre'] = pred
accuracy = model.evaluate(X_val, y_val)[1]
final_result.to_csv('DP_without_FeatureSelection{}.csv'.format(accuracy), index=0)
print('Finish! The predicted result has been saved as DP_without_FeatureSelection{}.csv'.format(accuracy))
```
